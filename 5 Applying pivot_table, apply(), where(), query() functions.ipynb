# Pivot Table 

#### 1) It quickly summarizes large amounts of data
#### 2) By default the aggregate function used is mean

##### syntax
pd.pivot_table(data,index,columns,values,aggfunc)

#### Q1) Using pivot_table create a dataframe displaying year wise mean of energy, length and popularity 

```python
pt1 = pd.pivot_table(df, index = ['year'], values = ['energy','length','popularity'])
pt1.head()
```

```python
r1 = df[df['year']==2018]
feat_mean = r1[['energy','length','popularity']].mean()
feat_mean
```

#### Q2) Using pivot_table create a dataframe displaying year wise max of energy, length and popularity where year is column

```python
pt2 = pd.pivot_table(df,columns = ['year'],
                     values = ['energy','length','popularity'],
                     aggfunc='max')
pt2.head(7)
```

```python
r2 = df[df['year']==2013]
r2[['energy','length','popularity']].max()
```

### Apply function on a series

#### Q1) Create a column named popularity bin. Generate the column values based on the following rules
a) It popularity <74, return low<br>
b) If popularity >=74 and popularity <=83, return medium<br>
c) If popularity >=83, return high

```python
def pop_bins(x):
    if x<74:
        return 'Low'
    elif x>=74 and x<83:
        return 'Medium'
    else:
        return 'High'
```

```python
df['Popularity_bins'] = df['popularity'].apply(pop_bins)
df[['popularity','Popularity_bins']].sample(10)
```

```python
df[['popularity','Popularity_bins']][df['popularity']<74]
```

# where()

It replace values where the condition is False

#### Syntax
df.where(condition, other)

where condition id True, keep the original value. Where False, replace with corresponding value from other. 

```python
df[df['year']==2018].shape
```

```python
df['Popularity_bins'].value_counts()
```

```python
# Replace the 
res = df['Popularity_bins'].where(df['Popularity_bins']!= 'Medium',"Med")
res
```

```python
print(res[:50])
```

```python
# Replace the energy value with high where energy is greater than mean energy
res2 = df['energy'].where(df['energy']<df['energy'].mean(),'High')
res2
```
