## Handling Null values in Ex2 

```python
d2 = pd.DataFrame({"a":[12,4,5,None,1],
                   "B":[None,2,54,3,None],
                   "C":[20,16,None,3,8],
                   "D":[14,3,None,np.NaN,6]})
ndf = pd.DataFrame(d2)
ndf.head()
```

```python
ndf1 = ndf.copy()
```

```python
ndf1.isnull().sum()
```

## Notes
<b>ffill<b> - Forward fill. Works column wise. Row i+1 is NaN, then ith Row values is forwarded and filled at (i+1)th row.
<b>bfill<b> - Backward fill. Works column wise. Row i is NaN, then i(i+1)th Row values is backwarded and filled at it row.

```python
ndf1
```

```python
ndf1.fillna(method='ffill')
```

```python
ndf1.fillna(method='bfill')
```

## Converting data types of columns

```python
store_data = {'Date':['2022-02-10','2022-02-11','2022-02-13','2022-02-14',
                      '2022-02-15','2022-02-16'],
              'Prod':['A','B','C','D','B','D'],
              'Sales':['4.59','8.37','6.5','5.82','7.23','9.0']}
store_df = pd.DataFrame(store_data)
store_df
```

```python
store_df.dtypes
```

```python
store_df['Date'] = pd.to_datetime(store_df['Date'])
store_df['Sales'] = pd.to_numeric(store_df['Sales'])
```

```python
store_df.dtypes
```

## Changing Index

```python
store_df.head()
```

```python
store_df.set_index('Prod',inplace = True)
store_df.head()
```
